## PIAZZA API REFRESH TOKEN FORM - http://10.61.64.32:8000/auth/token/refresh/
#####   Registers user to the server. Input should be in the format:
#####   {"refresh_token": "old_token"}

```python
# This is our PIAZZA API FORM TO REFRESH A TOKEN 
# 
#
#
#
# USER JASON FORMAT 
# ------------------------------------------------------------------------------------------------------------------------------------------
# old_token = "<old authentication token>" # <------------------  REPLACE THIS FIELD <old authentication token> WITH THE OLD TOKEN
# new_token = {"refresh_token": old_token}
# -----------------------------------------------------------------------------------------------------------------------------------------

# Importing the requests package for HTTP calls
import requests 

# The endopoint is=> http://10.61.64.32:8000/auth/token/refresh/
url = "http://10.61.64.32:8000/auth/token/refresh/"   

# ----------------------------------------------------------------------------------------------------------------------------------- 
# This will be our old token. We need to use this token to request a new one. (To refresh our token)

################################################################################################################################
old_token = "Or3ZYoND6CdIKnwjgMvuUn69fFWCVb" # <------------------  REPLACE THIS FIELD <old authentication token> WITH THE OLD TOKEN
################################################################################################################################


# Refresh your token.
new_token = {"refresh_token": old_token}  
# ------------------------------------------------------------------------------------------------------------------------------------

headers = {'Authorization': 'Bearer '+str(new_token)}
response = requests.post(url, headers = headers, data = new_token)
print(new_token)
```
